```python
def authenticate_and_get_client():
    """è™•ç† Google æˆæ¬Šä¸¦å–å¾— gspread å®¢æˆ¶ç«¯ã€‚"""
    global gc
    if gc is None:
        try:
            print("\nè«‹é»æ“Šé€£çµå®Œæˆ Google æˆæ¬Šï¼š")
            auth.authenticate_user()
            print("Google æˆæ¬Šå®Œæˆï¼")
            from google.auth import default
            creds, _ = default()
            gc = gspread.authorize(creds)
            print("gspread æˆæ¬ŠæˆåŠŸã€‚")
            return gc
        except Exception as e:
            print(f"âŒ æˆæ¬Šå¤±æ•—: {e}")
            return None
    else:
        return gc
gc = authenticate_and_get_client()
```

```python
def run_analysis(gc):
    """
    å¾ Google Sheets è®€å–è³‡æ–™ï¼ŒåŸ·è¡Œè¨ˆç®—ï¼Œä¸¦å›å‚³çµæœ DataFrame å’Œåˆ†ææ–‡å­—ã€‚
    """
    if gc is None:
        return pd.DataFrame(), pd.DataFrame(), "âŒ éŒ¯èª¤ï¼šgspread å°šæœªæˆæ¬Šã€‚è«‹å…ˆåŸ·è¡Œæ­¥é©ŸäºŒå®Œæˆæˆæ¬Šã€‚"

    try:
        sh = gc.open(spreadsheet_name)
        worksheet = sh.worksheet(worksheet_name)
        try:
            summary_ws = sh.worksheet("Summary")
        except gspread.WorksheetNotFound:
            summary_ws = sh.add_worksheet(title="Summary", rows="100", cols="20")

        print(f"è©¦ç®—è¡¨ '{spreadsheet_name}' é€£æ¥æˆåŠŸï¼Œé–‹å§‹è®€å–æ•¸æ“š...")
        data = worksheet.get_all_records()
        df = pd.DataFrame(data)

        if df.empty:
            return pd.DataFrame(), pd.DataFrame(), "âš ï¸ æ•¸æ“šåˆ†é ç‚ºç©ºï¼Œç„¡æ³•é€²è¡Œåˆ†æã€‚"

        df['é‡‘é¡'] = pd.to_numeric(df['é‡‘é¡'], errors='coerce')
        df.dropna(subset=['é‡‘é¡'], inplace=True)

        if df.empty:
            return pd.DataFrame(), pd.DataFrame(), "âš ï¸ æ•¸æ“šåˆ†é ä¸­ 'é‡‘é¡' æ¬„ä½ç„¡æœ‰æ•ˆæ•¸å­—ï¼Œç„¡æ³•é€²è¡Œåˆ†æã€‚"
        total_amount = df['é‡‘é¡'].sum()
        total_records = len(df)

        category_subtotals = df.groupby('åˆ†é¡')['é‡‘é¡'].sum().reset_index()
        category_subtotals.columns = ['åˆ†é¡', 'é‡‘é¡å°è¨ˆ']

        all_payers = df['ä»˜æ¬¾äºº'].unique()
        num_payers = len(all_payers)
        average_share = total_amount / num_payers if num_payers > 0 else 0

        payer_payments = df.groupby('ä»˜æ¬¾äºº')['é‡‘é¡'].sum().reindex(all_payers, fill_value=0).reset_index()
        payer_payments.columns = ['ä»˜æ¬¾äºº', 'å·²ä»˜ç¸½é¡']
        payer_payments['æ‡‰åˆ†æ”¤é‡‘é¡'] = average_share
        payer_payments['æ·¨é¡'] = payer_payments['å·²ä»˜ç¸½é¡'] - payer_payments['æ‡‰åˆ†æ”¤é‡‘é¡']

        payer_payments_gradio = payer_payments.copy()
        payer_payments_gradio[['å·²ä»˜ç¸½é¡', 'æ‡‰åˆ†æ”¤é‡‘é¡', 'æ·¨é¡']] = payer_payments_gradio[['å·²ä»˜ç¸½é¡', 'æ‡‰åˆ†æ”¤é‡‘é¡', 'æ·¨é¡']].applymap('{:,.2f}'.format)
        summary_ws.clear()
        summary_ws.update_cell(1, 1, "ç¸½çµå ±å‘Š")
        summary_ws.update_cell(2, 1, "ç¸½æ¶ˆè²»é¡")
        summary_ws.update_cell(2, 2, float(total_amount))
        summary_ws.update_cell(3, 1, "ç¸½ç­†æ•¸")
        summary_ws.update_cell(3, 2, total_records)

        summary_ws.update_cell(4, 1, "æ¯äººå¹³å‡æ‡‰åˆ†æ”¤")
        summary_ws.update_cell(5, 1, "åˆ†é¡å°è¨ˆ")
        subtotals_list = [category_subtotals.columns.tolist()] + category_subtotals.values.tolist()
        summary_ws.update(range_name=f'A6:B{6 + len(subtotals_list) - 1}', values=subtotals_list)

        summary_ws.update_cell(5, 4, "AA åˆ†æ”¤çµæœ")
        aa_list = [payer_payments.columns.tolist()] + payer_payments.values.tolist()
        aa_output_ws = payer_payments.copy()
        aa_list = [aa_output_ws.columns.tolist()] + aa_output_ws.values.tolist()
        data_rows = len(aa_list)
        aa_range = f'D8:G{8 + data_rows - 1}'


        summary_ws.update(range_name=aa_range, values=aa_list)

        print("âœ… çµæœå·²æˆåŠŸå¯«å…¥ Google Sheets çš„ Summary åˆ†é ï¼")


        analysis_text = f"ğŸ’¡ **æ—¥å¸¸æ”¯å‡ºåˆ†æå ±å‘Š** ğŸ’¡\n\n"
        analysis_text += f"**ç¸½æ¶ˆè²»é¡**: NT${total_amount:,.2f} (å…± {total_records} ç­†ç´€éŒ„)\n"
        analysis_text += f"**æ¯äººå¹³å‡æ‡‰åˆ†æ”¤**: NT${average_share:,.2f} (å…± {num_payers} äºº)\n"

        analysis_text += "\n--- **åˆ†é¡é«˜å…‰** ---\n"
        if not category_subtotals.empty:
            max_cat = category_subtotals.iloc[category_subtotals['é‡‘é¡å°è¨ˆ'].argmax()]
            min_cat = category_subtotals.iloc[category_subtotals['é‡‘é¡å°è¨ˆ'].argmin()]

            analysis_text += f"ğŸ’¸ **æœ€å¤§æ”¯å‡ºåˆ†é¡**: **{max_cat['åˆ†é¡']}**ï¼Œå…± NT${max_cat['é‡‘é¡å°è¨ˆ']:,.2f}ã€‚\n"
            analysis_text += f"ğŸŒ± **æœ€å°æ”¯å‡ºåˆ†é¡**: **{min_cat['åˆ†é¡']}**ï¼Œå…± NT${min_cat['é‡‘é¡å°è¨ˆ']:,.2f}ã€‚\n"
            analysis_text += f"\nğŸ‘‰ **å»ºè­°**: æª¢è¦– '{max_cat['åˆ†é¡']}' æ˜¯å¦æœ‰ç¯€çœç©ºé–“ã€‚\n"

        analysis_text += "\n--- **AA åˆ†æ”¤æ·¨é¡æ‘˜è¦** ---\n"

        should_receive = payer_payments[payer_payments['æ·¨é¡'] > 0].sort_values(by='æ·¨é¡', ascending=False)
        should_pay = payer_payments[payer_payments['æ·¨é¡'] < 0].sort_values(by='æ·¨é¡', ascending=True)

        if not should_receive.empty:
            receiver = should_receive.iloc[0]
            analysis_text += f"ğŸ’° **æœ€é«˜æ‡‰æ”¶æ¬¾**: **{receiver['ä»˜æ¬¾äºº']}**ï¼Œæ‡‰æ”¶ **{receiver['æ·¨é¡']:,.2f}** å…ƒã€‚\n"

        if not should_pay.empty:
            payer = should_pay.iloc[0]
            analysis_text += f"ğŸ“‰ **æœ€é«˜æ‡‰ä»˜æ¬¾**: **{payer['ä»˜æ¬¾äºº']}**ï¼Œæ‡‰ä»˜ **{-payer['æ·¨é¡']:,.2f}** å…ƒã€‚\n"

        analysis_text += "\n**è«‹åƒè€ƒä¸‹æ–¹çš„ AA åˆ†æ”¤çµæœè¡¨æ ¼ï¼Œé€²è¡Œçµç®—ã€‚**"

        return category_subtotals, payer_payments_gradio[['ä»˜æ¬¾äºº', 'å·²ä»˜ç¸½é¡', 'æ‡‰åˆ†æ”¤é‡‘é¡', 'æ·¨é¡']], analysis_text

    except gspread.WorksheetNotFound:
        return pd.DataFrame(), pd.DataFrame(), f"âŒ éŒ¯èª¤ï¼šæ‰¾ä¸åˆ°è³‡æ–™åˆ†é : {worksheet_name}ã€‚è«‹æª¢æŸ¥åˆ†é åç¨±ã€‚"
    except gspread.SpreadsheetNotFound:
        return pd.DataFrame(), pd.DataFrame(), f"âŒ éŒ¯èª¤ï¼šæ‰¾ä¸åˆ°è©¦ç®—è¡¨: {spreadsheet_name}ã€‚è«‹æª¢æŸ¥æª”æ¡ˆåç¨±ã€‚"
    except Exception as e:
        return pd.DataFrame(), pd.DataFrame(), f"âŒ ç™¼ç”ŸæœªçŸ¥éŒ¯èª¤: {e}"

def gradio_wrapper():
    """ç¢ºä¿æ¯æ¬¡éƒ½ä½¿ç”¨å…¨åŸŸçš„ gc å®¢æˆ¶ç«¯ä¾†å‘¼å«åˆ†æå‡½æ•¸ã€‚"""
    return run_analysis(gc)
```

```python
def gradio_wrapper():
    """ç¢ºä¿æ¯æ¬¡éƒ½ä½¿ç”¨å…¨åŸŸçš„ gc å®¢æˆ¶ç«¯ä¾†å‘¼å«åˆ†æå‡½æ•¸ã€‚"""
    global gc
    return run_analysis(gc)

with gr.Blocks(title="æ—¥å¸¸æ”¯å‡ºæ•¸ç®—èˆ‡åˆ†æ”¤å ±å‘Š") as demo:
    gr.Markdown("# ğŸ’¸ Google Sheets æ—¥å¸¸æ”¯å‡ºæ•¸ç®—èˆ‡åˆ†æ”¤å ±å‘Š ğŸ“Š")
    ai_analysis_output = gr.Markdown(label="AI åˆ†æçµæœ", value="é»æ“Šä¸‹æ–¹æŒ‰éˆ•é–‹å§‹åˆ†æ...")
    run_button = gr.Button("é»æ“Šæˆ‘æ›´æ–°ä¸¦åŸ·è¡Œåˆ†æ")

    gr.Markdown("---")
    gr.Markdown("## ğŸ“‹ è¨ˆç®—è©³ç´°çµæœ")

    with gr.Row():
        category_df_output = gr.DataFrame(
            headers=["åˆ†é¡", "é‡‘é¡å°è¨ˆ"],
            label="åˆ†é¡æ”¯å‡ºå°è¨ˆ"
        )
        aa_df_output = gr.DataFrame(
            headers=["ä»˜æ¬¾äºº", "å·²ä»˜ç¸½é¡", "æ‡‰åˆ†æ”¤é‡‘é¡", "æ·¨é¡"],
            label="AA åˆ†æ”¤æ·¨é¡ (+è™Ÿç‚ºæ‡‰æ”¶, -è™Ÿç‚ºæ‡‰ä»˜)"
        )

    # 5. å®šç¾©äº’å‹•é‚è¼¯ (æŒ‰éˆ•é»æ“Šè§¸ç™¼åˆ†æ)
    run_button.click(
        fn=gradio_wrapper,
        inputs=None,
        outputs=[category_df_output, aa_df_output, ai_analysis_output]
    )

    # 6. åœ¨é é¢åŠ è¼‰æ™‚è‡ªå‹•åŸ·è¡Œä¸€æ¬¡åˆ†æ
    demo.load(
        fn=gradio_wrapper,
        inputs=None,
        outputs=[category_df_output, aa_df_output, ai_analysis_output]
    )

# 7. å•Ÿå‹• Gradio ä»‹é¢
print("\n--- å•Ÿå‹• Gradio ä»‹é¢ ---")
demo.launch(debug=True, share=True)
```
